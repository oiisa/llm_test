```python
import spacy
spacy.cli.download("ru_core_news_sm")
nlp = spacy.load("ru_core_news_sm")
```

```python
city = []
for message in data.message:
  doc = nlp(message)
  # print(doc.ents)
  # city_message = ""
  # for entity in doc.ents:
  #   # print(entity.text, sep=' ')
  #   if entity.label_ == "LOC":
  #     city_message = entity.text
  #   if len(city_message) == 0:
  #     city_message = None
  if len(doc.ents) != 0:
    city.append(doc.ents[0].text)
  else:
    city.append(None)
```

```python
data["city"] = pd.Series(city)
data
```

```python
data.to_csv("data.csv")
```

```python
for m in data[data["city"].isna()].message:
  print(nlp(m.capitalize()).ents)
```
